```python
import pandas as pd
import numpy as np

# read in all our data
nobel_data = pd.read_csv("archive.csv")

# set seed for reproducibility
np.random.seed(0) 
```

```python
# Look at a few rows of the nfl_data file. A handful of missing data!
nobel_data.sample(4)
```

```python
# get the number of missing data points per column
missing_values_count_nobel = nobel_data.isnull().sum()

# look at the # of missing points in the first ten columns
print(missing_values_count_nobel)
```

```python
# How many total missing values do we have?
# shape returns the dimentionality of a dataframe (rows and columns), can you guess what product will do?
total_cells_nobel = np.product(nobel_data.shape) 
total_missing_nobel = missing_values_count_nobel.sum()

# percent of data that is missing
percentage_missign_values_nobel = (total_missing_nobel/total_cells_nobel) * 100
print(percentage_missign_values_nobel)
```

```python
in_nobel = nobel_data[nobel_data["Laureate Type"] == "Individual"]
in_nobel
```

```python
org_nobel = nobel_data[nobel_data["Laureate Type"] == "Organization"]
org_nobel
```

```python
#Fixing Laureate Type which assigned nobel prizes for individual as an Organization
fix_nobel_data =  nobel_data.loc[(nobel_data["Laureate Type"] == "Organization") &  (nobel_data["Birth Date"].notnull()) ]
fix_nobel_data = fix_nobel_data.replace("Organization", "Individual")
fix_nobel_data
```

```python
final_ind_nobel = in_nobel.append(fix_nobel_data)
final_ind_nobel
```
