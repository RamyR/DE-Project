```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

nobel = pd.read_csv("archive.csv")

# half_count = len(nobel) / 2
# nobel = nobel.dropna(thresh=half_count,axis=1) # Drop any column with more than 50% missing values

nobel
```

```python
org_nobel = nobel[nobel["Laureate Type"] == "Organization"]
in_nobel = nobel[nobel["Laureate Type"] == "Individual"]



g = org_nobel.loc[(org_nobel['Birth Date'].notnull())]
# g= w['Laureate Type'].map({'female': 1, 'male': 0})

s =  g.replace("Organization", "Individual")



new_indu =  in_nobel.append(s)

# new_indu['Birth Date'] = pd.to_datetime(new_indu['Birth Date'], format='%Y-%m-%d')

new_indu
```

```python
new_org = org_nobel.loc[(org_nobel["Laureate Type"]== "Organization") & (org_nobel['Birth Date'].isna())] 

new_org = new_org.drop(['Birth Date','Birth City','Birth Country','Sex','Death Date','Death Country','Death City','Organization City','Organization Name','Organization Country'], axis=1)
new_org
```

```python

#in the individual dataset USA is the dominante in winning the nobel prizes
from matplotlib.pyplot import hist


u = new_indu['Birth Country'].value_counts().rename_axis('Countries').reset_index(name='Number_of_Nobel_Prizes')

u = u [0:10]
u
```

```python
sns.set()
# and setting the size of all plots.
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [11, 7]

# Plotting USA born winners 
ax = sns.lineplot(data=u, x="Countries", y="Number_of_Nobel_Prizes")
# ax.set_xticklabels(u['Countries'], rotation='vertical', fontsize=10)
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='large'  
)
plt.show()


# # Adding %-formatting to the y-axis
# from matplotlib.ticker import PercentFormatter
# ax.yaxis.set_major_formatter(PercentFormatter())
```

```python
# Calculating the proportion of USA born winners per decade
new_indu['usa_born_winner'] = new_indu["Birth Country"] == "United States of America"
new_indu['decade'] = (np.floor(new_indu["Year"] / 10) * 10).astype(int)
prop_usa_winners = new_indu.groupby("decade", as_index=False)["usa_born_winner"].mean()

prop_usa_winners
```

```python
plt.rcParams['figure.figsize'] = [11, 7]

ax = sns.lineplot(x=prop_usa_winners["decade"], y=prop_usa_winners["usa_born_winner"])

# Adding %-formatting to the y-axis
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter())
```
